### Free form exercise

The following mock-up dataframe summarizes the actual and forecasted ice-cream consumption (in tons) for four different cities in Switzerland. Create a line plot that represents the yearly evolution of consumption using a different color for each city. Include a graph title, a legend with a correct title, and rename the axes if needed. Make sure the consumption axis starts at 0 tons and ends at 10 tons, using intervals of 2 tons along the axis. 

Brownie points: overlay a scatter plot on top of your line plot, where the point size is defined as the ratio of the forecasted consumption to the actual consumption. Add a legend with a correct title to account for this new attribute. Hint: define the new aesthetics directly in `geom_point()`

```r
library(ggplot2)
library(dplyr)
library(tidyr)
```

```r
head(ice_cream)
```

```r
ggplot(ice_cream, aes(year, consumption, color=city)) + 
    geom_line() + 
    geom_point(aes(size=forecast/consumption)) + 
    ylab("Ice-cream consumption (tons)") +
    xlab("Year") +
    ggtitle("Ice-cream consumption 2013-2017") + 
    scale_colour_discrete(name="City", l=50) + 
    scale_size_continuous(name="Forecast / actual ratio") + 
    scale_y_continuous(breaks = seq(0, 10, 2), limits = c(0, 10)) +
    theme_bw()
```
